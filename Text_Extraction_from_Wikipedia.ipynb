```
import nltk
import urllib
import bs4 as bs
import re
import requests
from nltk.corpus import stopwords
nltk.download('stopwords')

nltk.download('punkt')

nltk.download('averaged_perceptron_tagger')
```

```
#Getting the data from wikipedia
import requests 
from bs4 import BeautifulSoup 
  
URL = "https://en.wikipedia.org/wiki/Thor"
r = requests.get(URL) 
  
soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib 
print(soup.prettify())
```

```
source = urllib.request.urlopen("https://en.wikipedia.org/wiki/Thor").read()

soup1 = bs.BeautifulSoup(source, "lxml")

soup1
```

```
#Feteching the data

text = ""

for paragraph in soup.findAll('p'):
  text += paragraph.text
```

```
text
```

```
#preprocessing the text data

import string

text = text.lower()
text = "".join([char for char in text if char not in string.punctuation])
print(text)
```

```
#Removing numbers
text = re.sub(r'\d+', '', text)

print(text)
```

```
#Removing white spaces

text = text.strip()

print(text)
```

```
text = re.sub(r'\[[0-9]*\]', '', text)

print(text)
```

```
sentences = nltk.sent_tokenize(text)

sentences
```

```
#Building triple extractor which takes the sentence 

from nltk.tokenize import word_tokenize, sent_tokenize 
stop_words = set(stopwords.words('english')) 

# sent_tokenize is one of instances of 
# PunktSentenceTokenizer from the nltk.tokenize.punkt module 


for i in sentences: 
	
	# Word tokenizers is used to find the words 
	# and punctuation in a string 
	wordsList = nltk.word_tokenize(i) 

	# removing stop words from wordList 
	wordsList = [w for w in wordsList if not w in stop_words] 

	# Using a Tagger. Which is part-of-speech 
	# tagger or POS-tagger. 
	tagged = nltk.pos_tag(wordsList) 

	print(tagged) 
```

```
tagged = nltk.pos_tag(sentences) 

print(tagged) 
```
